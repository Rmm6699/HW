```python
import matplotlib.pyplot as plt
import cv2
from retinaface import RetinaFace
```

```python
img_path = "/content/drive/MyDrive/3/test.jpg"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
faces = RetinaFace.extract_faces(img_path = img_path, align = False)
for face in faces:
  plt.imshow(face)
  plt.show()
```

```python
faces = RetinaFace.detect_faces(img_path)
faces
```

```python
def find_face(img,bbox_dict,landmark=False):
    key = list(bbox_dict.keys())
    for i in key:
        identity = faces[i]
        facial_area = identity["facial_area"]
        img = cv2.rectangle(img, (facial_area[2], facial_area[3]), (facial_area[0], facial_area[1]), (255, 255, 255), 2)
        if landmark:
            landmarks = identity["landmarks"]
            cv2.circle(img, tuple(map(int, landmarks["left_eye"])), 2,(0, 0, 255), 2)
            cv2.circle(img, tuple(map(int, landmarks["right_eye"])), 2,(0, 0, 255), 2)
            cv2.circle(img, tuple(map(int, landmarks["nose"])), 2,(0, 0, 255), 2)
            cv2.circle(img, tuple(map(int, landmarks["mouth_left"])), 2,(0, 0, 255), 2)
            cv2.circle(img, tuple(map(int, landmarks["mouth_right"])), 2,(0, 0, 255), 2)
    return img
```

```python
img2 = img.copy()
find_face(img2,faces,landmark=False)
plt.imshow(img2)
```

```python
img2 = img.copy()
find_face(img2,faces,landmark=True)
plt.imshow(img2)
```
